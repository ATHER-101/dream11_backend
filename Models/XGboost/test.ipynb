```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
all_data=pd.read_csv('../../Data/updated_consolidated_dataset.csv')

opponent_batting_cols = [col for col in all_data.columns if col.startswith('opponentplayer') and 'bowling_points' in col]
all_data['opponent_bowling_strength'] = all_data[opponent_batting_cols].mean(axis=1)

current_batting_cols= [col for col in all_data.columns if col.startswith('player') and 'batting_points' in col]
all_data['current_batting_strength'] = all_data[current_batting_cols].mean(axis=1)


avg_all_data=pd.read_csv('../../Data/avg_updated_consolidated_dataset.csv')

opponent_batting_cols = [col for col in avg_all_data.columns if col.startswith('opponentplayer') and 'bowling_points' in col]
avg_all_data['opponent_bowling_strength'] = avg_all_data[opponent_batting_cols].mean(axis=1)

current_batting_cols= [col for col in avg_all_data.columns if col.startswith('player') and 'batting_points' in col]
avg_all_data['current_batting_strength'] = avg_all_data[current_batting_cols].mean(axis=1)
```

```python
# print(all_data['venue'].unique())
print(all_data['venue'].nunique())
print(all_data['venue'].value_counts())

```

```python
X_bat=all_data[['venue', 'strike_rate', 'wickets', 'maidens',
       'economy_rate', 'bowled_lbw', 'catches', 'stumpings',
       'batting_points_avg','venue_avg', 'variance','opponent_bowling_strength','current_batting_strength' ]]
# 'player1_batting_points','player2_batting_points','player3_batting_points','player4_batting_points','player5_batting_points','player6_batting_points','player7_batting_points','player8_batting_points','player9_batting_points','player10_batting_points','opponentplayer1_bowling_points','opponentplayer2_bowling_points','opponentplayer3_bowling_points','opponentplayer4_bowling_points','opponentplayer5_bowling_points','opponentplayer6_bowling_points','opponentplayer7_bowling_points','opponentplayer8_bowling_points','opponentplayer9_bowling_points','opponentplayer10_bowling_points','opponentplayer11_bowling_points'
y_bat=all_data['batting_points']
```

```python
encoder = OneHotEncoder(sparse_output=False)  # sparse_output=False gives a dense array

# Fit and transform the venue column
encoded = encoder.fit_transform(X_bat[['venue']])

# Create a DataFrame for the one-hot encoded columns
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(['venue']))

# Concatenate the one-hot encoded columns with the original DataFrame
X_bat = pd.concat([X_bat.drop(columns=['venue']), encoded_df], axis=1)
# X_bat.drop(columns=['venue'],inplace=True)
```

```python
X_bat.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_bat, y_bat, test_size=0.3, random_state=40)
from xgboost import XGBRegressor
model=XGBRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test,y_pred))
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
print(y_pred[1:100])
print(y_test[1:100])
```
