```python
import os
import pandas as pd

def find_outliers_in_folder(folder_path, column_name):
    # Iterate through all CSV files in the folder
    for file_name in os.listdir(folder_path):
        if file_name.endswith(".csv"):  # Ensure it's a CSV file
            file_path = os.path.join(folder_path, file_name)
            
            try:
                # Load the CSV file
                df = pd.read_csv(file_path)
                
                # Identify rows where the column values are not strings
                non_string_rows = df[~df[column_name].apply(lambda x: isinstance(x, str))]
                
                if not non_string_rows.empty:
                    print(f"Outliers found in file '{file_name}':")
                    print(non_string_rows)
                    print("-" * 50)
                # else:
                #     print(f"No outliers in file '{file_name}'.")
            except Exception as e:
                print(f"Error processing file '{file_name}': {e}")

# Example usage
folder_path = "C:/Users/itsme/OneDrive/Documents/GitHub/dream11_backend/Data/Datasets/ipl_json_processed"  # Replace with the path to your folder containing CSVs
column_name = "player_id"  # Replace with the name of the column to check

find_outliers_in_folder(folder_path, column_name)
```
